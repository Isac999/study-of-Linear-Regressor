```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x, y = make_regression(n_samples=200, n_features=1, noise=25)
plt.scatter(x, y)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
lr = LinearRegression().fit(x_train, y_train)
print('Teste: {:.2f}% \nTreino: {:.2f}%'.format((lr.score(x_test, y_test))*100, (lr.score(x_train, y_train))*100))
```

```python
plt.scatter(x_test, y_test, color='red', label='y_Values')
plt.plot(x_test, lr.predict(x_test), color='blue', label='Predict')
plt.title('Train Score: {:.1f}% \nTest Score: {:.1f}%'.format((lr.score(x_train, y_train)*100), (lr.score(x_test, y_test)*100)))
plt.xlabel('Input Data (X)')
plt.ylabel('Output Values (Y)')
plt.legend()
plt.show()
```
